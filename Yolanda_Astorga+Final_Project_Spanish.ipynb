```python
datos.isnull().sum().sort_values(ascending=True)
```

#### Los valores nulos serán imputados mediante un CICLO FOR utilizando los siguientes criterios:
- cálculo de la MODA para cada columna con DATOS de TIPO CATEGÓRICO
- cálculo de la MEDIA para cada columna con DATOS de TIPO NUMÉRICO

```python
for col in datos.columns:
    if datos[col].dtype=='O': 
        print("Rellenando valores nulos en la columna categórica:",col)
        datos[col].fillna(value=datos[col].mode()[0], inplace=True)
    else:
        print("Rellenando valores nulos en la columna numérica:", col)
        datos[col].fillna(value=datos[col].mean(), inplace=True)
```

### Tarea 5b.1. ¿El precio del servicio y el precio de la habitación tienen un impacto mutuo? Ilustre esta relación con un gráfico de dispersión e indique sus inferencias

```python
plt.rcParams['legend.fontsize'] = 'large'
plt.figure(figsize=(12, 10))

sns.scatterplot(data= datos, x='service_fee', y='price', hue='room_type', s=150, palette= 'Set1')

plt.xticks(fontsize= 16)
plt.yticks(fontsize=16)
plt.xlabel('Tarifa del Servicio', fontsize=14)
plt.ylabel('Precio de la Habitación', fontsize=14)

plt.legend(title='Tipo Habitación', loc= 'center right', bbox_to_anchor= (0.90 , 0.50 ))

plt.title('Impacto entre el Precio de Habitación y la Tarifa de Servicio', fontsize= 16, y= 1.03)

plt.show()
```

```python
correlation= datos['service_fee'].corr(datos['price'])
print('Coeficiente de correlación:', correlation)
```

- A la vista de este gráfico de dispersión no cabe duda del impacto mutuo existente entre el Precio de la Habitación y la Tarifa del Servicio.
  
-  Claramente, podemos observar cómo ambos valores tienden a incrementarse juntos. Esto no quiere decir que uno sea causa o consecuencia del otro, pues dicha tendendcia puede ser debida a otros factores no analizados aquí.
  
-  Técnicamente, este impacto mutuo o asociación directa entre estos valores, se define como "Correlación Lineal Positiva Casi Perfecta".

---
CONCLUSIÓN: Existe una fuerte relación entre el precio de la habitación y la tarifa del servicio: ambos aumentan de forma directa, independientemente de cuál sea el precio de la habitación o la tarifa del servicio.

### Tarea 5b.2. Utilizando un gráfico lineal muestre en qué año tuvo lugar la máxima construcción de habitaciones.

```python
lugares_por_año = datos['construction_year'].value_counts().sort_index()
print(lugares_por_año)
```

```python
año_máximo = lugares_por_año.idxmax()

plt.figure(figsize=(12, 6))
sns.lineplot(x=lugares_por_año.index, y=lugares_por_año.values, marker='o', color= 'b')

plt.scatter(año_máximo, lugares_por_año[año_máximo], color='red', label=f'Máxima construcción en {año_máximo}')
plt.annotate(f'Máxima construcción en {año_máximo}', xy=(año_máximo, lugares_por_año[año_máximo]),
              xytext=(año_máximo + 2, lugares_por_año[año_máximo] + 10),
            arrowprops=dict(facecolor= 'red', shrink=0.05))

plt.title('Número de Habitaciones Construidas por Año',fontsize= 16, y= 1.03)
plt.xlim(2002.8, 2022.2)
plt.xticks(range(2003,2023))
plt.xlabel('Año de construcción', fontsize= 14)
plt.ylabel('Número de Habitaciones', fontsize= 14)
plt.legend(loc= 'center left')
plt.grid(True)
plt.show()
```
